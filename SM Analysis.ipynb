```python
import matplotlib.pyplot as plt

labels = ['1-2 hrs', '2-4 hrs', '>4 hrs', '<1 hr']
sizes = df['How many hours on average in a day do you spend on social media?'].value_counts()

explode = (0.2,0.2,0.2,0.2)
colors = ["#61D199","#857E7B","#FCD757","#FBBA72"]

fig1, ax1 = plt.subplots()

ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90, colors=colors, pctdistance=0.85)
ax1.axis('equal')

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['1-2 hrs', '2-4 hrs', '>4 hrs', '<1 hr']
sizes = df['How many hours on average in a day do you spend on social media?'].value_counts()

labels_prod = ['Agree','Disagree','Neutral','Strongly disagree','Strongly agree','Agree','Disagree','Neutral','Strongly disagree','Strongly agree','Agree','Disagree','Neutral','Strongly disagree','Strongly agree','Agree','Disagree','Neutral','Strongly disagree','Strongly agree']
sizes_prod = df['I notice that my productivity has diminished due to social media.'].value_counts()

colors = ["#61D199","#857E7B","#FCD757","#FBBA72"]
colors_prod = ["#ED9390","#B5E2FA","#5B6057","#EDDEA4","#EFBCD5"]

fig1, ax1 = plt.subplots()

ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90, colors=colors, pctdistance=0.9, frame=True)
ax1.pie(sizes_prod, autopct='%1.1f%%',radius=0.75,
        shadow=False, startangle=90, colors=colors_prod, pctdistance=0.5)
ax1.axis('equal')

centre_circle = plt.Circle((0,0),0.40,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.tight_layout()
plt.show()
```
